```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data = pd.read_csv("titanic.csv")
data
```

```python
data.head(10)
```

```python
data.tail(5)
```

```python
input=data.drop('Survived',axis='columns')
input
```

```python
target=data['Survived']
target
```

```python
la_name =LabelEncoder()
la_gender	=LabelEncoder()
la_ticket=LabelEncoder()
la_cabin=LabelEncoder()
la_embarked=LabelEncoder()
```

```python
input['name_n']=la_name.fit_transform(input['Name'])
input['gender_n']=la_gender.fit_transform(input['Gender'])
input['ticket_n']=la_ticket.fit_transform(input['Ticket'])
input['cabin_n']=la_cabin.fit_transform(input['Cabin'])
input['embarked_n']=la_embarked.fit_transform(input['Embarked'])
```

```python
input
```

```python
input_n=input.drop(['Name', 'Gender', 'Ticket', 'Cabin', 'Embarked'],axis='columns')
input_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```
